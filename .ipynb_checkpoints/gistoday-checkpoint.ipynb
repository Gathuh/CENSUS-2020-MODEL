# Reading the data set from the computer and preprocessing
Here's a summary of the key information discussed regarding the use of pandas for manipulating DataFrames:

### Key Functions and Methods in Pandas

1. **`df.columns`**:
   - Returns the column labels of the DataFrame as an `Index` object.
   - Allows you to access or modify column names.

2. **`df.iloc[0]`**:
   - Selects the first row of the DataFrame based on its integer position (0-based indexing).
   - Returns the row as a pandas Series.

3. **`data.drop(index=0)`**:
   - Removes the row at index position 0 from the DataFrame.
   - By default, it returns a new DataFrame without the specified row. To modify in place, use `inplace=True`.

4. **Dropping Columns**:
   - To drop a column, use `df.drop('column_name', axis=1)`:
     - `'column_name'`: The name of the column to drop.
     - `axis=1`: Specifies that a column is being dropped.
   - You can drop multiple columns by passing a list, e.g., `df.drop(['col1', 'col2'], axis=1)`.
   - To drop by index position, use `df.drop(df.columns[column_index], axis=1)`.
   - To modify the DataFrame in place, use `inplace=True`.

### Additional Notes
- After dropping a row or column, you may want to reset the index using `df.reset_index(drop=True)` to maintain a clean, consecutive index.
- `.iloc[]` can also be used for more complex indexing, including selecting specific rows and columns based on their positions.

This summary encapsulates the key operations and functionalities discussed regarding DataFrame manipulation in pandas, including accessing, dropping, and modifying data. Let me know if you need any further details or clarifications!

## Calculate population density

```python

data["Total"] = pd.to_numeric(data["Total"], errors='coerce')
data["Male"] = pd.to_numeric(data["Male"], errors='coerce')
data["Female"] = pd.to_numeric(data["Female"], errors='coerce')
data["SQ.Km"] = pd.to_numeric(data["SQ.Km"], errors='coerce')

data=data.astype({"Total":float,"Male":float,"Female":float,"SQ.Km":float})
data["Population_desity"]=data["Total"]/data["SQ.Km"]
data.info()
```

## ADD coordinates using opencage API

```python
from opencage.geocoder import OpenCageGeocode
import time
API_KEY = '8da62f1ba72e4ca486ad7fac09ace958'
geocoder = OpenCageGeocode(API_KEY)
```

```python


df = pd.DataFrame(data)

# Function to get coordinates
def get_coordinates(subcounty):
    query = f"{subcounty}, Kenya"
    result = geocoder.geocode(query)
    
    if result:
        return result[0]['geometry']['lat'], result[0]['geometry']['lng']
    else:
        return None, None

# Add longitude and latitude columns
latitudes = []
longitudes = []

for subcounty in df.iloc[:,1]:
    lat, lng = get_coordinates(subcounty)
    latitudes.append(lat)
    longitudes.append(lng)
    
    # Optional: wait a bit to avoid hitting API rate limits
    time.sleep(1)

# Adding new columns to DataFrame
df['Latitude'] = latitudes
df['Longitude'] = longitudes

# Display the updated DataFrame
print(df)
```

##  MAPS
###  Choropleth Map or a Point Density Map,

```python
import folium
map_center = [-1.286389, 36.817223]  # Coordinates for Nairobi
m = folium.Map(location=map_center, zoom_start=7)

# Add points to the map
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=(row['Latitude'], row['Longitude']),
        radius=row['Population_desity'] / 5000,  # Adjust the radius for better visibility
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"{row['subcounty']}: {row['Population_desity']:.2f} people/sq km",
    ).add_to(m)

# Save the map to an HTML file
m.save("/home/gathu/Documents/work/GIS/kenya_population_density_map2.html")
```

```python
import folium
from folium.plugins import HeatMap

# Coordinates for Nairobi
map_center = [-1.286389, 36.817223]

# Create a base map
m = folium.Map(location=map_center, zoom_start=7)

# Prepare data for the heat map
heat_data = []

# Loop through the DataFrame and append latitude, longitude, and population density to heat_data
for idx, row in df.iterrows():
    # Append [latitude, longitude, weight] for the heat map
    heat_data.append([row['Latitude'], row['Longitude'], row['Population_desity']])

# Create the heat map
HeatMap(heat_data, radius=15, blur=10, max_zoom=1).add_to(m)

# Save the map to an HTML file
m.save("/home/gathu/Documents/work/GIS/kenya_population_density_heatmap.html")
```
